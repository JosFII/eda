```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
data=pd.read_excel('result.xlsx')
```

Data berisi data dari semua normal monster di yugioh pada saat ini

```python
data
```

```python
mean=np.nanmean(data['atk'])
print("mean:",mean)
median=np.nanmedian(data['atk'])
print("median:",median)
var=np.nanvar(data['atk'])
print("variance:",var)
stdev=np.nanstd(data['atk'])
print("standar deviasi:",stdev)
mode=stats.mode(data['atk'],nan_policy='omit')
print("mode:",mode.mode)
```

```python
plt.hist(data['atk'],bins=10,alpha=0.7,color='#ff772c',edgecolor='#3acaf2')
plt.title('histogram nilai atk')
plt.xlabel('atk')
plt.ylabel('fequensi')
plt.show()
```

Berdasarkan bentuk dari histogram, ini merupakan distribusi left skew atau negatif skew
