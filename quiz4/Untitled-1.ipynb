```python
data = np.array([1, 2, 2, 3, 4, 5, 6])

mean = np.mean(data)

median = np.median(data)

mode = stats.mode(data)

variance = np.var(data)

std_dev = np.std(data)

print(f"Mean: {mean}")

print(f"Median: {median}")

print(f"Mode: {mode.mode}")

print(f"Variance: {variance}")

print(f"Standard Deviation: {std_dev}")
```

```python
plt.hist(data, bins=20)

plt.title('Histogram Distribusi Data')

plt.xlabel('Nilai')

plt.ylabel('Frekuensi')

plt.show()
```

```python
plt.boxplot(data)

plt.title('Boxplot Distribusi Data')

plt.show()
```

```python
stats.probplot(data, dist="norm", plot=plt)

plt.title('QQ-plot')

plt.show()
```

```python

stat, p_value = stats.shapiro(data)

print(f'Statistik Uji: {stat}, p-value: {p_value}')

if p_value > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

```python
stat, p_value = stats.kstest(data, 'norm')

print(f'Statistik Uji: {stat}, p-value: {p_value}')


if p_value > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

```python
result = stats.anderson(data)

print(f'Statistik Uji: {result.statistic}')

for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print(f'Pada tingkat signifikansi {sl}%, data mengikuti distribusi normal')
    else:
        print(f'Pada tingkat signifikansi {sl}%, data tidak mengikuti distribusi normal')
```
