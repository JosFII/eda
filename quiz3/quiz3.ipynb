```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data=pd.read_excel('phhy.xlsx')
```

```python
data
```

```python
plt.hist(data['def'],bins=10,color='red',edgecolor='#00a3d9')
plt.title('histrogram untuk nilai defence pada kartu di set photon hypernova',color='#c7a801')
plt.xlabel('defence',color='#00a3d9')
plt.ylabel('frekuensi',color='#ff0000')
plt.show()
```

```python
plt.scatter(data['atk'],data['level/link val'],color='#f7e9b3')
plt.title('atk vs level / link value')
plt.xlabel('atk')
plt.ylabel('level vs link val')
plt.show()
```

```python
jumlahattri=data['attribute'].value_counts()
```

```python
plt.pie(jumlahattri,labels=jumlahattri.index,autopct='%1.0f%%')
plt.title('distribusi attribute')
plt.show(True)
```

```python
jumlah_arce=data['archetype'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.bar(jumlah_arce.index,jumlah_arce,color='#ff97c3')
plt.title('jumlah kartu dalam archetype')
plt.xlabel('nama archetyppe')
plt.ylabel('jumlah kartu')
plt.show()
```
