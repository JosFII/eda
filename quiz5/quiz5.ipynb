```python
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
```

Data yang digunakan dalah data karakter dari game limbus company, data merupakan saat karakter lvl 45 dan uptie 4, data diambil dari [limbuscompany.wiki.gg](limbuscompany.wiki.gg)

```python
limbus=pd.read_excel("limbus1.xlsx")
limbus
```

variabel yang akan digunakan adalah sebagai berikut:
- Max HP= HP maximum yang dimiliki character
- Min speed= speed minimum yang dimiliki karakter sebelum faktor lain
- max speed= speed maximum yang dimiliki karakter sebelum faktor lain
- def level = nilai level dari denfense yang karakter miliki
- off level 1 =nilai dari offense level pada skill 1 karakter
- off level 2 =nilai dari offense level pada skill 2 karakter
- off level 3 =nilai dari offense level pada skill 3 karakter (jika karakter memiliki skill 3 - alternatif, maka nilai dari yang alternatif yang akan digunakan)

```python
used_data=limbus[["max HP","min speed","max speed","def level","off level 1","off level 2","off level 3"]]
used_data
```

```python
covariance=used_data.cov()
covariance
```

```python
correlation=used_data.corr()
correlation
```

```python
sea.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title("heatmap matriks corelasi")
plt.show()
```

```python
sea.heatmap(correlation,annot=True,cmap='Blues')
plt.title("heatmap matriks corelasi")
plt.show()
```

```python
sea.pairplot(used_data)
plt.title("pairplot data")
plt.show()
```
